```python
import scipy.optimize as sco
import numpy as np
import pylab as py
```

```python
def fx(x):
    v = 2*(x[0]**2) + 30
    return v

def fx1(x):
    return 2*(x**2) + 30

x=np.linspace(-20,20,100)

y1=fx1(x)
py.plot(x,y1)

fopt= sco.fmin(fx,x,full_output=True,disp=True)
```
