```python
%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("data_for_3d_graph.csv")
```

```python
df = df[df.year == 2018].drop('year',axis=1)
```

```python
df = df.groupby(['month','object'],as_index=False).sum()
```

```python
fr = df['object'].unique()
```

```python
df['object'].replace(fr,np.arange(len(fr)) + 1,inplace=True)
```

```python
fr.tolist()
```

```python
fig = plt.figure(figsize=(8, 3))
ax1 = fig.add_subplot(111, projection='3d')


x = df['month'].values
y = df['object'].values
z = np.zeros_like(df['amount'].values)

dx = np.ones_like(df['month'].values)
dy = np.ones_like(df['object'].values)
dz = df['amount'].values

ax1.w_yaxis.set_ticklabels(fr)

ax1.bar3d(x, y, z, dx, dy, dz, shade=True)
```
